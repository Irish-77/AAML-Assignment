```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

from IPython.display import display
from PIL import Image
from PIL.ImageFilter import BoxBlur
```

```python
image = Image.open('test.jfif')
display(image)
```

```python
convolution_kernel = np.array([[0, -1, 0], 
                               [-1, 5, -1], 
                               [0, -1, 0]])
```

```python
img = np.asarray(image)
result = cv2.filter2D(img, -1, kernel = convolution_kernel)
display(Image.fromarray(result.astype(dtype = np.uint8), 'RGB'))
```
